## Set Up Diffusion Data ##

This script loads in the dataset and formats it for the tabddpm pipeline



```python
import sys

import numpy as np

from HullParameterization import Hull_Parameterization as HP

import csv

```

```python
#Set up variables

DesVecName = 'Input_Vectors.csv'

CwName = 'GeometricMeasures/Cw.csv'

SAName = 'GeometricMeasures/Area_WS.csv'

VolName = 'GeometricMeasures/Volume.csv'

DCName = 'GeometricMeasures/GaussianCurvature.csv'

MBName = 'GeometricMeasures/MaxBox.csv'

# NEEED TO SET PATH TO DATASET
#==============================================================================
DS_path = './Hull_DataSet/'

folder_root = 'Constrained_Randomized_Set_'

path = DS_path + folder_root
#==============================================================================

DesVec = []

CwVec = []

SAVec = []

VolVec = []

DCVec = []

MBVec = []

labels = []

for i in range(1,4):
        
    #Location of Design Vectors
    with open(path + str(i) + '/' + DesVecName) as csvfile:
        reader = csv.reader(csvfile)
        for count, row in enumerate(reader):
            if count != 0:
                DesVec.append(row)
    
    #Location of Cw Vectors
    with open(path + str(i) + '/' + CwName) as csvfile:
        reader = csv.reader(csvfile)
        for count, row in enumerate(reader):
            if count != 0:
                CwVec.append(row)
            else:
                labels = np.array(row)

                
                
    #Location of SA Vectors
    with open(path + str(i) + '/' + SAName) as csvfile:
        reader = csv.reader(csvfile)
        for count, row in enumerate(reader):
            if count != 0:
                SAVec.append(row)
            else:
                labels = np.array(row)
                
    #Location of Vol Vectors
    with open(path + str(i) + '/' + VolName) as csvfile:
        reader = csv.reader(csvfile)
        for count, row in enumerate(reader):
            if count != 0:
                VolVec.append(row)
            else:
                labels = np.array(row)
                
        #Location of DC Vectors
    with open(path + str(i) + '/' + DCName) as csvfile:
        reader = csv.reader(csvfile)
        for count, row in enumerate(reader):
            if count != 0:
                DCVec.append(row)
            else:
                labels = np.array(row)
                
                
        #Location of MB Vectors
    with open(path + str(i) + '/' + MBName) as csvfile:
        reader = csv.reader(csvfile)
        for count, row in enumerate(reader):
            if count != 0:
                MBVec.append(row)
            else:
                labels = np.array(row)
            
#Save as a np.float array
DesVec = np.array(DesVec)

DesVec = DesVec.astype(np.float32())

print(np.shape(DesVec)) #Should be (300000,45)
print(DesVec[0])

#Save as a np.float array
CwVec = np.array(CwVec)
SAVec = np.array(SAVec)
VolVec = np.array(VolVec)
DCVec = np.array(DCVec)
MBVec = np.array(MBVec)


CwVec = CwVec.astype(np.float32())
SAVec = SAVec.astype(np.float32())
VolVec = VolVec.astype(np.float32())
DCVec = DCVec.astype(np.float32())
MBVec = MBVec.astype(np.float32())
```

```python
# Now lets clean up X

idx_BBFactors = [33,34,35,36,37]
idx_BB = 31

idx_SBFactors = [38,39,40,41,42,43,44]
idx_SB = 32

for i in range(0,len(DesVec)):
    
    DesVec[i,idx_BBFactors] = DesVec[i,idx_BB] * DesVec[i,idx_BBFactors] 
    DesVec[i,idx_SBFactors] = DesVec[i,idx_SB] * DesVec[i,idx_SBFactors]

X = np.copy(DesVec[:,1:])

# Should be (30000,44)
print(np.shape(X))
print(X[0])
```

```python
#Lets add the constraints to the vector too
cons = np.zeros((len(DesVec),49))


for i in range(0,len(DesVec)):
    hull = HP(DesVec[i])
    cons[i] = hull.input_Constraints()

print(cons.shape)

print(cons[0])
cons = cons.astype(np.float32())
```

```python
# Now Clean Up Each Performance Metric:

# All of these performance metrics are log scaled so that they have a normal distribution

Y_Cw= np.sum(np.log10(CwVec),axis=-1)

Y_Cw = Y_Cw[:,np.newaxis]

print(Y_Cw.shape)


Y_SA = np.log(SAVec[:,[4,9]])
print(Y_SA.shape)

Y_Vol = -np.log(VolVec[:,[4,9]])  # We want negative volume to reward large volume when we 'minimize' it later

print(Y_Vol.shape)

Y_MB = -MBVec[:,4]/VolVec[:,9] # We want negative max box volume to reward large volume when we 'minimize' it later
Y_MB = Y_MB[:,np.newaxis]
print(Y_MB.shape)

Y_DC = np.log10(DCVec)
#Y_DC = Y_DC[:,np.newaxis]
print(Y_DC.shape)





```

```python
# If it all looks good, lets concatenate it all together

Y = np.concatenate((Y_Cw,Y_SA, Y_Vol, Y_MB, Y_DC), axis=-1)

print(Y.shape)

#B = np.concatenate((Y[:,1:],cons,X),axis=-1) #MultiObj with constraints
#B = np.concatenate((Y[:,1:],X),axis=-1) #MultiObj without constraints.
B = np.copy(X) #Cw without constriants
#B = np.concatenate((cons,X),axis=-1) #Cw with cons

print(B.shape) #(30000,98)
```
